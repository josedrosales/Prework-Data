#  **Listas en Python**


Una lista es una colección de elementos ordenados que puede contener datos de diferentes tipos. Es mutable, lo que significa que sus elementos pueden cambiarse después de ser creada.

# Acceso a elementos de una lista


Puedes acceder a un elemento utilizando su posición (índice).
- El índice comienza en 0 para el primer elemento.
- Los índices negativos acceden desde el final hacia el principio.

```python
frutas = ["manzana", "pera", "plátano", "cereza", "uva", "kiwi", "naranja", "sandía", "fresa", "mango"]

# Acceder a elementos
print(frutas[0])  # Primer elemento
print(frutas[1])  #Segundo elemento
print(frutas[-1]) # Último elemento
```

```python
# Modificar elementos
edades = [2, 45, 60]
edades[1] = 2000  # Cambiar el segundo elemento
print(edades)
```

# Slice
Un slice es una forma de extraer una parte de una lista usando la notación lista [inicio:fin:paso]

inicio → índice donde empieza (incluido).

fin → índice donde termina (excluido).

paso → de cuánto en cuánto avanza (por defecto es 1).

```python
numeros = [10, 20, 30, 40, 50, 60]

print(numeros[1:4])   # [20, 30, 40] → desde índice 1 hasta 3
print(numeros[:3])    # [10, 20, 30] → desde el inicio hasta índice 2
print(numeros[3:])    # [40, 50, 60] → desde índice 3 hasta el final
print(numeros[::2])   # [10, 30, 50] → saltando de 2 en 2
```

# Usando índices negativos
Los índices negativos cuentan desde el final:

```python
print(numeros[-3:])   # [40, 50, 60] → últimos 3 elementos
print(numeros[:-2])   # [10, 20, 30, 40] → todo menos los últimos 2
```

Invertir una lista con slice

```python
print(numeros[::-1])  # [60, 50, 40, 30, 20, 10]
```

Modificar con slice

```python
numeros[1:3] = [99, 100]
print(numeros)  # [10, 99, 100, 40, 50, 60]
```

Insertar con slice

```python
numeros = [10, 20, 30, 40]
numeros[2:2] = [25, 26]
print(numeros)  # [10, 20, 25, 26, 30, 40]
```

Insertar y reemplazar

```python
numeros = [10, 20, 30, 40]
numeros[1:3] = [99, 100, 101]
print(numeros)  # [10, 99, 100, 101, 40]
```

# Iteración sobre listas

Iterar sobre una lista significa recorrer sus elementos uno por uno, permitiendo ejecutar operaciones sobre ellos. 
Esto se logra utilizando bucles como `for`, que nos permite recorrer toda la lista de manera sencilla.

```python
# Usando un bucle for
for i in frutas:
    print(i)
```

```python
# Contar cuántas palabras tienen más de 5 letras
contador = 0
for palabra in frutas:
    if len(palabra) > 5:
        contador += 1
print(f"Hay {contador} palabras con más de 5 letras.")
```

```python
# Uso de range()
for i in range(0, 10, 2): # range(start, stop, step) genera números desde start hasta stop saltando 2
    print(i)
```

```python
# Simulación de lanzamiento de un dado
import random

suma = 0
for i in range(6):
    dado = random.randint(1, 6)
    print(f"El dado ha lanzado: {dado}")
    suma += dado
print(f"La suma total es: {suma}")
```
